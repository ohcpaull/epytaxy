# Rigaku .RAS file plotting

```python
from vlabs.xray import ras_file
import numpy as np
import matplotlib.pyplot as plt
```

```python
xrd = ras_file("data/P0142OP_2theta_15-115deg_83mins.ras")
```

```python
xrd
```

```python
fig, ax = plt.subplots()

ax.plot(xrd["2theta"], xrd["counts"])
ax.set(yscale="log", xlabel="2theta-omega", ylabel="Intensity (a.u.)")
```
